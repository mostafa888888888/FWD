```python
data.drop(['PatientId','AppointmentID','AppointmentDay','ScheduledDay'],axis=1,inplace=True)
data.head()
```

```python
data.rename(columns={'No-show':'NO_SHOW'},inplace=True)
data.rename(columns={'Hipertension':'Hypertension'},inplace=True)
data.head()
```

```python
data.hist(figsize=(17,14));
```

```python
# most of patients suffer from Hypertension
# most of patients arenot enroll in ascholarship
# 33% of patients received sms
```

```python
show=data['NO_SHOW']=='No'
noshow=data['NO_SHOW']=='Yes'
```

```python
data[show].count()
```

```python
data[noshow].count()
```

```python
# 25% of Patients didnot show
```

```python
plt.figure(figsize=[14,8])
data.Gender[show].hist(alpha=0.5,label='show')         
data.Gender[noshow].hist(alpha=0.5,label='noshow')         
plt.legend()
plt.title('the relation between gender And show' )
plt.xlabel('GENDER')
plt.ylabel('patients number');
```

```python
# no relation
```

```python
plt.figure(figsize=[14,8])
data.Scholarship[show].hist(alpha=0.5,label='show')         
data.Scholarship[noshow].hist(alpha=0.5,label='noshow')         
plt.legend()
plt.title('the relation between Scholarship And show' )
plt.xlabel('GENDER')
plt.ylabel('patients number');
```

```python
# no relation
```

```python
plt.figure(figsize=[14,8])
data.Hypertension[show].hist(alpha=0.5,label='show')         
data.Hypertension[noshow].hist(alpha=0.5,label='noshow')         
plt.legend()
plt.title('the relation between Hypertension And show' )
plt.xlabel('GENDER')
plt.ylabel('patients number');
```

```python
# no relation
```

```python
plt.figure(figsize=[14,8])
data.SMS_received[show].hist(alpha=0.5,label='show')         
data.SMS_received[noshow].hist(alpha=0.5,label='noshow')         
plt.legend()
plt.title('the relation between SMS_received  And show' )
plt.xlabel('GENDER')
plt.ylabel('patients number');
```

```python
# alot of Patients show without receiving sms
```

```python
plt.figure(figsize=[14,8])
data.Neighbourhood[show].value_counts().plot(kind='bar',alpha=0.5,label='show')         
data.Neighbourhood[noshow].value_counts().plot(kind='bar',alpha=0.5,color='green',label='noshow')         
plt.legend()
plt.title('the relation between Neighbourhood  And show' )
plt.xlabel('GENDER')
plt.ylabel('patients number');
```
